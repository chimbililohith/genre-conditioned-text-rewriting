```python
df_1 = df.rename(columns = {"genre" : "source_style",
                              "text" : "raw_content"})
df_1 = df_1.drop("id", axis=1)
print(df_1.head())
```

```python
df_1.to_csv('/content/drive/MyDrive/evaluation_dataset_400.csv', index=False)
```

```python
import math
import random
import pandas as pd

# 1. Load your dataset (update path if needed)
df = df_1
# 2. Get list of genres from the 'source_style' column
genres = sorted(df["source_style"].unique().tolist())

# 3. For reproducibility
random.seed(42)

# 4. Create balanced target_style for each source genre
df["target_style"] = None  # initialize column

for g in genres:
    # Indices for rows with this source genre
    idx = df.index[df["source_style"] == g].tolist()
    n = len(idx)

    # Possible target genres (cannot map to itself)
    possible_targets = [x for x in genres if x != g]  # 5 other genres

    # We want to distribute n rows as evenly as possible over these 5 targets
    # Build a list like [g2, g3, g4, g5, g6, g2, g3, ...] to reach length n
    repeats_needed = math.ceil(n / len(possible_targets))
    target_list = (possible_targets * repeats_needed)[:n]

    # Shuffle so target genres are mixed within this source genre
    random.shuffle(target_list)

    # Assign to target_style for these rows
    df.loc[idx, "target_style"] = target_list

# 5. check distribution
print(df.groupby(["source_style", "target_style"]).size().unstack(fill_value=0))

# 6. Save new CSV
df.to_csv("/content/drive/MyDrive/external_genre_validation_400_with_targets.csv", index=False)
print("Saved to external_genre_validation_400_with_targets.csv")
```
