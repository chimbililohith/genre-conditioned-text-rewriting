```python
def fetch_gutenberg_text(gid: int) -> str | None:
    # most common patterns, in order
    patterns = [
        f"https://www.gutenberg.org/files/{gid}/{gid}-0.txt",
        f"https://www.gutenberg.org/files/{gid}/{gid}.txt.utf-8",
        f"https://www.gutenberg.org/files/{gid}/{gid}.txt",
        f"https://www.gutenberg.org/cache/epub/{gid}/pg{gid}.txt",
    ]
    for url in patterns:
        try:
            resp = requests.get(url, timeout=30)
            if resp.status_code == 200 and len(resp.text.strip()) > 200:
                return resp.text
        except Exception:
            pass
    return None
```

```python
os.makedirs(BASE_DIR, exist_ok=True)

for genre, books in GENRE_BOOKS.items():
    genre_dir = os.path.join(BASE_DIR, genre)
    os.makedirs(genre_dir, exist_ok=True)

    for filename, gid in books.items():
        out_path = os.path.join(genre_dir, f"{filename}.txt")
        if os.path.exists(out_path):
            print(f"[skip] {out_path} already exists")
            continue

        print(f"[downloading] {genre}/{filename}  (Gutenberg #{gid})")
        text = fetch_gutenberg_text(gid)
        if text is None:
            print(f"  -> FAILED to download {gid}")
            continue

        with open(out_path, "w", encoding="utf-8") as f:
            f.write(text)
        print(f"  -> saved to {out_path}")

print("✅ done.")
```
